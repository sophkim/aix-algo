#### 실전 문제 2. 숫자 카드 게임: 여러 개의 숫자 카드 중에서 가장 높은 숫자가 쓰인 카드 한 장을 뽑는 게임. 이때 뽑고자 하는 행에서 가장 낮은 숫자를 뽑아야 하는 것이 조건

- 입력 조건
    첫째 줄에 숫자 카드들이 놓인 행의 개수 N과 열의 개수 M이 공백을 기준으로 자연수로 주어짐 (1<=N, M<=100)\
    둘째 줄부터 N개의 줄에 걸쳐 각 카드에 적힌 숫자가 주어짐, 각 숫자는 1이상 10,000 이하의 자연수\


- 출력 조건: 첫째 줄에 게임의 룰에 맞게 선택한 카드에 적힌 숫자를 출력

```python
### 내 풀이 -> list sort 사용 최솟값

data = '''3 3
3 1 2
4 1 4
2 2 2
'''

n, m = map(int, data.splitlines()[0].split())
max_row_item = 0 
for i, line in enumerate(data.splitlines()[1:]):
    l_to_int = list(map(int, line.split()))
    l_to_int.sort()
    if i == 0:
        max_row_item = l_to_int[0]
    else:
        if l_to_int[0] > max_row_item:
            max_row_item = l_to_int[0]
print(max_row_item)
```

```python
### 답지 풀이 -> min 활용 

data = '''3 3
3 1 2
4 1 4
2 2 2
'''
# n=3, m=3, 1st row는 메타데이터
n, m = map(int, data.splitlines()[0].split())
data = [list(map(int, line.split())) for line in data.splitlines()[1:]]
result = 0
for row in data:
    min_value = min(row)
    result = max(result, min_value)
print(result)
```

### ✏️ Notes

#### 1. **nlogn** 복잡도: nlogn은 입력 크기 n이 커질 때, 실행 시간이 n에 비례하면서 동시에 logn 만큼 추가로 계산 복잡도가 증가함을 의미 
nlogn의 원인: 대부분의 경우, 데이터를 반복적으로 반으로 쪼개고, 각 단계에서 전체 데이터를 한 번씩 처리=> 즉 (1)+(2)=(3) \
(1) logn: 분할 단계 수, 데이터를 2로 계속 나누는 횟수
예:n→n/2→n/4→…→1, 단계 수 = log2n 

(2) n: 각 단계에서의 작업량
각 단계마다 전체 원소를 한 번씩 처리 -> 즉, 단계당 O(n) 

(3) 결합: 총 작업량=O(n)×O(logn)=O(nlogn) 

<p style='font-weight:bold; color:orange'> nlogn 요약 </p> : nlogn = 분할 깊이 × 전체 처리량 (쪼개고, 전부 훑는 처리량) 
            (ex) 학생 n명이 친 시험의 수 logn, 총 채점량: nlogn

#### 2. **Input()** 입력 함수 정리: 다음 줄을 하나 소비하는 함수(Iterator의 next()처럼 기능)

#### 3. 내 접근법 VS 풀이 코드 => 최솟값만 필요 시 List sort()[0] 보다 min() 활용할 것!

| 항목        | `min()`    | `sort() + [0]` |
| --------- | ---------- | -------------- |
| 시간 복잡도    | O(n)       | O(n log n)     |
| 비교 횟수     | n−1        | 훨씬 많음          |
| 메모리       | 추가 없음      | 내부 스택 사용       |
| 원본 리스트 변경 | ❌          | ✅              |
| 목적 적합성    | 최솟값만 필요할 때 | ❌              |
