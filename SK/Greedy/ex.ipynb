### Greedy Algorithm: 현재 상황에서 가장 좋은 것 고르기, 최적의 해 찾기 
예제 1. 동전을 최소로 쓰면서 거스름 돈 거슬러 주기 

```python
# 예제 1. 동전을 최소로 쓰면서 거스름 돈 거슬러 주기
## 문제: 동전 500 원, 100원, 50원, 10원 존재, 거슬러 줘야 하는 돈이 N원일 때, 동전의 최소 개수 구하기

N = 1260 # 거슬러 줘야 하는 돈
count = 0 

coin_types = [10, 100, 50, 500]
coin_types.sort(reverse=True) # 큰 단위의 동전부터 사용하기 위해 내림차순 정렬

for c in coin_types:
    count += N//c # 해당 동전으로 나눠지면 count 증가
    N -= (N//c) * c 

print(f'사용된 동전의 개수: {count}')
```
