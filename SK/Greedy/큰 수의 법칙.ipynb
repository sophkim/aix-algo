### Greedy Algorithm: 현재 상황에서 가장 좋은 것 고르기, 최적의 해 찾기 

#### 실전 문제 1. 큰 수의 법칙: 다양한 수로 이루어진 배열이 있을 때 주어진 수열을 M번 더하여 가장 큰 수를 만드는 법칙, 이때 특정 수를 연속해서 K번 초과하여 더할 수 없음

- 입력 조건
    첫째 줄에 N(2<=N<=1,000), M(1<=M<=10,000),K(1<=K<=10,000)의 자연수가 주어지며, 자연수는 공백으로 구분 \
    둘째 줄에 N개의 자연수 O, 자연수는 공백으로 구분, 각각의 자연수는 1 이상 10,000 이하 \
    입력으로 주어지는 K는 항상 M보다 작거나 같음

- 출력 조건: 첫째 줄에 큰 수의 법칙에 따라 더해진 답을 출력

```python

# # input 있는 상황 -> sys 모듈 사용, input()보다 빠름
# import sys
# data = sys.stdin.readline().rsstrip()

## 수열로 풀기

data = """5 8 3
2 4 5 4 6"""

lines = data.splitlines()
N, M, K = map(int, lines[0].split()) # 공백 기준으로 첫 째줄 분리 
array = list(map(int, lines[1].split())) # 공백 기준으로 둘 째줄 분리
array.sort(reverse=True) # 내림차순 정렬

top, second = array[0], array[1]
seq = [top] * (K) + [second] # 수열 만들기 
seq_len = len(seq)
repeated_seq = M // seq_len # 반복할 수 있는 수열 수 
plus_sec = M % seq_len 
total = sum(seq) * repeated_seq + sum((second for _ in range(plus_sec)))
print(total)

```

```python
data = """5 8 3
2 4 5 4 6"""

lines = data.splitlines()
N, M, K = map(int, lines[0].split()) # 공백 기준으로 첫 째줄 분리 
array = list(map(int, lines[1].split())) # 공백 기준으로 둘 째줄 분리
array.sort(reverse=True) # 내림차순 정렬
top_num = M//K 
total_top_num = top_num * K
sec_num = M - total_top_num
total = (array[0] * total_top_num) + (array[1] * sec_num)
print(total)
```
