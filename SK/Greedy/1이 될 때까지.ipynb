### Greedy Algorithm: 현재 상황에서 가장 좋은 것 고르기, 최적의 해 찾기 

#### 실전 문제 3. N이 1이 될 때까지의 최소 연산 횟수: 어떠한 수 N이 1이 될 때까지 N에서 1을 빼고, N을 K로 나누는 두 가지 과정을 반복. 단 나누기는 N이 K로 나누어 질 때만 가능

- 입력 조건: 첫째 줄에 N(2<=N<=100,000),K(2<=K<=100,000)가 공백으로 구분된 자연수로 주어짐. N>=K

- 출력 조건: 첫째 줄에 N이 1이 될 때까지 수행하는 연산 횟수의 최솟값을 출력

```python
### 내 풀이 -> 1씩 빼기 

data = ''' 25
5
'''

N, K = map(int, data.splitlines())
print(f'초기 자연수 N: {N}, 나누는 수 K: {K}')
count = 0 
div = 0 
sub = 0 
while True:
    if N ==1:
        break
    elif N%K==0:
        N = N//K
        div +=1
    else:
        N -=1
        sub += 1
count = div + sub
print(f'최소 연산 횟수: {count}, 나누기 연산 횟수: {div}, 빼기 연산 횟수: {sub}')
```

```python
### 정답 풀이; 한 번에 많이 빼기

data = ''' 11119
13
'''

N, K = map(int, data.splitlines())
print(f'초기 자연수 N: {N}, 나누는 수 K: {K}')
result = 0 
while True:
    # N이 정확히 나누어 떨어지는 조건 X, 가능한 div 연산 수 세기에 초점
    target = N//K *K
    result += (N-target) # 나머지 
    N = target 
    if N<K: #나머지가 K보다 작은 순간, 즉 나누기가 불가능해 질 때
        break
    result +=1
    N //= K

result += N-1
print(result)
```

### ✏️ Notes

#### **Point** : target = (N // K) * K, N에 근접한 K의 배수를 빼줌으로써 코드 실행 시 뺄셈 횟수 반복 줄이기 

target = (N // K) * K \
result += (N - target) \
N = target
