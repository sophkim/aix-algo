### DFS&BFS Algorithm: 탐색 알고리즘 

#### 실전 문제 3. 음료수 얼려 먹기: 상하 좌우가 붙어 있는 경우(1), 붙어 있는 것으로 간주할 때 총 생성되는 아이스크림 개수는? 

- 입력 조건
1. 첫째 줄에 세로 길이 N과 가로 길이 M이 주어진다.(1<=N, M<=1,000)
2. 두 번째 줄부터 N+1 줄까지 얼음 틀의 형태가 주어진다.
3. 이때 구멍이 뚫려있는 부분은 0, 그렇지 않은 부분은 1이다. 

- 출력 조건: 한 번에 만들 수 있는 아이스크림의 개수를 출력

```python
# Idea: 상하좌우 탐색 후 모두 1로 막힌 경우 아이스크림 개수 +1(True), 현재 위치에서 상하좌우에서 재귀 탐색 => DFS!

input = '''3 3
  0 0 1 
  0 1 0  
  1 0 1
  '''
n, m = map(int, input.splitlines()[0].split())
graph = []

# 데이터 리스트 형태 그래프로 입력 
for i in range(n):
    line_data = list(map(int, input.splitlines()[i+1].split()))
    graph.append(line_data)

def dfs(x, y):
    if x < 0 or x >n-1 or y < 0 or y > m-1:
        return False
    if graph[x][y] == 0: 
        graph[x][y] = 1   #방문 처리
        # 현재 위치에서 연결된 상하좌우 탐색
        dfs(x-1, y)
        dfs(x+1, y)
        dfs(x, y-1)
        dfs(x, y+1)
        return True

    return False   #1이 등장하거나 구간을 벗어나면 False 

result = 0
for i in range(n):
    for j in range(m):
        if dfs(i, j) == True:
            result +=1

print(result)
```
