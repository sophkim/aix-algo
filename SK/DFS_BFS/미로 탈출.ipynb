### DFS&BFS Algorithm: 탐색 알고리즘 

#### 실전 문제 4. 미로 탈출: 현재 위치는 (1,1)이고 미로의 출구는 (N,M)에 위치. 괴물이 없는 부분은 1로 표시 되었을 때, 탈출을 위해 움직여야 하는 최소 칸의 수

- 입력 조건
1. 첫째 줄에 두 정수 N, M이 주어진다.(4<=N, M<=200)
2. 다음 N개의 줄에는 각각 M개의 정수(0혹은 1)로 미로의 정보가 주어진다. 각각의 수는 공백 없이 제시된다. 
3. 시작과 마지막 칸은 항상 1이다. 

- 출력 조건: 첫째 줄에 최소 이동 칸의 개수를 출력

```python
# Idea: 목적지까지의 최단 거리 탐색->깊이 탐색하는 것보다 인접한 구간 탐색 후 나아가는 것이 best, BFS
### Detailed approach: 1. 인접 노드 큐 삽입 2. 큐 꺼내고 상하좌우 확인 3. 이동 가능하면 현재 위치 +1, 즉 이동한 위치를 누적하며 방문하지 않은 곳만 최단 거리로 인식

from collections import deque

data = """ 3 3
1 1 0
1 1 1
0 1 1
"""

N, M = map(int, data.splitlines()[0].split())

graph = [] 
for i in range(1, N+1):
    graph.append(list(map(int, data.splitlines()[i].split())))

# 이동할 수 있는 좌표 
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

# 큐가 빌 때까지 반복
def bfs(x, y, cond=True):
  queue = deque()
  queue.append((x, y)) # 시작 (0,0) 입력 
  
  while queue: 
    x, y = queue.popleft() # 현재 위치 pop
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if nx <0 or ny<0 or nx>N-1 or ny>M-1: 
            continue
        if graph[nx][ny]==0:
            continue
        if graph[nx][ny]==1: #이동 가능 (지나온 곳은 1보다 커짐)
            graph[nx][ny] = graph[x][y] + 1 
            if cond==True and nx == N-1 and ny == M-1: # 목적지 도달 시 이미 최단 경로 도달이므로 더이상 다른 경로 탐색 X 
                return graph[nx][ny]
            queue.append((nx, ny)) # 큐 삽입
  return graph[N-1][M-1]

print(bfs(0, 0, cond=True))
```
